```python
import numpy as np
import pandas as pd
import scipy.stats as sp
import matplotlib.pyplot as plt 
```

```python
# this function plot probability distribution 

def plot_probability (numbers):
    snumbers=pd.Series(numbers).round(2)
    total=len(numbers)
    n=snumbers.value_counts().sort_index()
    probability = n / total
    plt.figure(figsize=(10, 5))
    plt.style.use('dark_background')
    plt.grid(alpha=0.1)
    plt.scatter(n.index,probability,color='#a6fd29')  # this realy calculate probability 
    # plt.hist(numbers, bins=50, density=True, alpha=1, color='skyblue', edgecolor='black') # this one show distribution
    print(f"Entropy is : {-sum(probability * np.log(probability))}")
```

```python

numbers=np.random.uniform(0,1,1000000)
plot_probability(numbers)
```

```python
numbers=np.random.normal(0,0.1,1000000)
plot_probability(numbers)
```

```python
numbers=np.random.poisson(5,1000000)
plot_probability(numbers)
```
